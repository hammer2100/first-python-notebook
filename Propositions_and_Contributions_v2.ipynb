## Knight Center for Journalism: Analyzing California Public Data

```python
import pandas as pd
```

```python
props = pd.read_csv("http://www.firstpythonnotebook.org/_static/committees.csv")
props.head()
```

```python
props.info()
```

```python
contribs = pd.read_csv("http://www.firstpythonnotebook.org/_static/contributions.csv")
contribs.head()
```

```python
contribs.info()
```

```python
props.prop_name.value_counts()
```

#### Looking at Prop 64

```python
prop64 = props[props.prop_name == "PROPOSITION 064- MARIJUANA LEGALIZATION. INITIATIVE STATUTE."]
```

```python
prop64.head()
```

```python
prop64.info()
```

```python
prop64_contribs = pd.merge(prop64, contribs, on="calaccess_committee_id")
```

```python
prop64_contribs.head()
```

```python
prop64_contribs.info()
```

```python
prop64_contribs.amount.sum()
```

```python
prop64_contribs.committee_position.value_counts()
```

```python
support_amount = prop64_contribs[prop64_contribs.committee_position == "SUPPORT"].amount.sum()
print(support_amount)
```

```python
oppose_amount = prop64_contribs[prop64_contribs.committee_position == "OPPOSE"].amount.sum()
print(oppose_amount)
```

```python
import numpy as np
pd.pivot_table(prop64_contribs, values=["amount"], columns=["committee_position"], aggfunc=np.sum)
```

```python
total_contribs = prop64_contribs.amount.sum()
print(total_contribs)
prop64_contribs["percentage_total"] = prop64_contribs["amount"]/(prop64_contribs["amount"].sum())
print(prop64_contribs["percentage_total"].sum())
```

```python
pd.pivot_table(prop64_contribs, values=["percentage_total"], columns=["committee_position"], aggfunc=np.sum)
```

```python
prop64_contribs["perc_count"] = 1/prop64_contribs.committee_name_x.count()
prop64_contribs["perc_count"].head()
```

```python
pd.pivot_table(prop64_contribs, values=["perc_count"], columns=["committee_position"],aggfunc=np.sum)
```

```python
prop64_contribs.sort_values("amount", ascending=False)
```

```python
prop64_contribs[prop64_contribs.committee_position == "OPPOSE"].sort_values("amount", ascending = False)
```

```python
prop64_contribs[prop64_contribs.committee_position == "SUPPORT"].groupby("committee_name_x").amount.sum().reset_index().sort_values("amount", ascending=False)
```

```python
prop64_contribs.groupby(["contributor_firstname","contributor_lastname","committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```
